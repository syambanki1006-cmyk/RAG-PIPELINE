```python
from langchain_core.documents import Document

from google.colab import files

# Upload file from your PC
uploaded = files.upload()

documents = []

for filename in uploaded.keys():
    content = uploaded[filename].decode("utf-8")
    documents.append(Document(page_content=content, metadata={"source": filename}))

print(documents)
```

```python
word_count = len(documents[0].page_content.split())

print(word_count)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

fixed_splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=0
)

m1 = fixed_splitter.split_documents(documents)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

sliding_splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20
)

m2 = sliding_splitter.split_documents(documents)

```

```python
!pip install sentence-transformers
from langchain_experimental.text_splitter import SemanticChunker
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

semantic_splitter = SemanticChunker(embeddings)

m3 = semantic_splitter.split_documents(documents)
```

```python
print("Method 1 Chunks\n")

for i, chunk in enumerate(m1):
    print(f"Chunk {i+1}")
    print(chunk.page_content)
    print("-"*50)

print("\nMethod 2 Chunks\n")

for i, chunk in enumerate(m2):
    print(f"Chunk {i+1}")
    print(chunk.page_content)
    print("-"*50)

print("\nMethod 3 Chunks\n")

for i, chunk in enumerate(m3):
    print(f"Chunk {i+1}")
    print(chunk.page_content)
    print("-"*50)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

emb_m1 = HuggingFaceEmbeddings(
    model_name="Alibaba-NLP/gte-Qwen2-1.5B-instruct"
)

vectors_m1 = emb_m1.embed_documents([doc.page_content for doc in m1])

print(len(vectors_m1))
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

emb_m2 = HuggingFaceEmbeddings(
    model_name="intfloat/e5-large-v2"
)

vectors_m2 = emb_m2.embed_documents([doc.page_content for doc in m2])

print(len(vectors_m2))
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

emb_m3 = HuggingFaceEmbeddings(
    model_name="BAAI/bge-m3"
)

vectors_m3 = emb_m3.embed_documents([doc.page_content for doc in m3])

print(len(vectors_m3))
```

```python
import os
os.environ["CHROMA_API_KEY"] = "**************************************************"
os.environ["CHROMA_TENANT"] = "****************************************"
os.environ["CHROMA_DATABASE"] = "**********"
```

```python
import os
import chromadb
from langchain_chroma import Chroma

# Read credentials from environment variables
api_key = os.getenv("CHROMA_API_KEY")
tenant = os.getenv("CHROMA_TENANT")
database = os.getenv("CHROMA_DATABASE")

# Connect to Chroma Cloud
client = chromadb.CloudClient(
    api_key=api_key,
    tenant=tenant,
    database=database
)

# Create / connect collection
vectordb = Chroma(
    client=client,
    collection_name="m1_collection",
    embedding_function=emb_m1
)

# Insert chunks
vectordb.add_documents(m1)

print("Inserted m1 chunks into Chroma Cloud")
```

```python
import os
from pinecone import Pinecone
from langchain_pinecone import PineconeVectorStore

# Initialize Pinecone
pc = Pinecone(api_key=os.getenv("PINECONE_API_KEY"))

index_name = "student"

# Connect to index
index = pc.Index(index_name)

# Create vector store
vectordb_pinecone = PineconeVectorStore(
    index=index,
    embedding=emb_m2
)

# Insert m2 chunks
vectordb_pinecone.add_documents(m2)

print("m2 chunks inserted into Pinecone")
```

```python
import os

os.environ["WEAVIATE_URL"] = "**********************************************.weaviate.cloud"
os.environ["WEAVIATE_API_KEY"] = "*********************************************************************"
```

```python
import os
import weaviate
from langchain_weaviate import WeaviateVectorStore

# connect to Weaviate
client = weaviate.connect_to_weaviate_cloud(
    cluster_url=os.getenv("WEAVIATE_URL"),
    auth_credentials=weaviate.auth.AuthApiKey(os.getenv("WEAVIATE_API_KEY"))
)

# create vector store
vectordb_weaviate = WeaviateVectorStore(
    client=client,
    index_name="M3Collection",
    text_key="text",
    embedding=emb_m3
)

# insert documents
vectordb_weaviate.add_documents(m3)

print("m3 chunks inserted into Weaviate")
```
