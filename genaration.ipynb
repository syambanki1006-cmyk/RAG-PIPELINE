```python
from typing import List
from langchain_core.documents import Document

def build_augmented_prompt(query: str, docs: List[Document]):

    if not docs:
        return "No context available."

    context = "\n\n".join(
        [f"[Chunk {i+1}] {doc.page_content}" for i, doc in enumerate(docs)]
    )

    prompt = f"""
        "You are an expert document explainer.\n"
        "Rules:\n"
        "- Use ONLY the information in the CONTEXT.\n"
        "- Do NOT add outside knowledge.\n"
        "- Do NOT infer or guess.\n"
        "- If the answer is not explicitly stated, say:\n"
        "  'The provided records do not mention this information.'"

CONTEXT:
{context}

QUESTION:
{query}

ANSWER:
"""
    return prompt
```

```python
def query_groq(prompt):

    response = client.responses.create(
        model="openai/gpt-oss-120b",
        input=prompt
    )

    return response.output_text
```

```python
def generate_answer(query, retriever):

    docs = retriever.invoke(query)   # Pinecone retriever

    prompt = build_augmented_prompt(query, docs)

    answer = query_groq(prompt)

    print("\n===== Pinecone Answer =====\n")
    print(answer)

```

```python
query = "how  Atorvastatin should used "

generate_answer(query, mmr_retriever)
```

```python
def generate_answer_chroma(query, chroma_vectorstore):

    # Cosine Similarity Retrieval
    retriever = chroma_vectorstore.as_retriever(
        search_type="similarity",
        search_kwargs={"k": 4}
    )

    docs = retriever.invoke(query)

    prompt = build_augmented_prompt(query, docs)

    answer = query_groq(prompt)

    print("\n===== ChromaDB (Cosine Similarity) =====\n")
    print(answer)

```

```python
query = "how  Atorvastatin should used "

generate_answer_chroma(query, vectordb_chroma)
```

```python
def generate_answer_weaviate(query, weaviate_vectorstore):

    retriever = weaviate_vectorstore.as_retriever(
        search_type="similarity", # Changed from "hybrid" to "similarity"
        search_kwargs={
            "k": 4,
            "alpha": 0.5 # Alpha parameter still enables hybrid search in Weaviate
        }
    )

    docs = retriever.invoke(query)

    prompt = build_augmented_prompt(query, docs)

    answer = query_groq(prompt)

    print("\n===== Weaviate (Hybrid Search) =====\n")
    print(answer)

query = "how  Atorvastatin should used."

generate_answer_weaviate(query, vectordb_weaviate)
```
